```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Airbnb NYC 2019.csv')

# --- Data Exploration (Optional, but good for understanding the data) ---
print("First 5 rows of the dataset:")
print(df.head())

print("\nInformation about the dataset:")
print(df.info())

# --- Data Visualizations ---

# 1. Distribution of Airbnb Listing Prices
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Distribution of Airbnb Listing Prices in NYC')
plt.xlabel('Price (USD)')
plt.ylabel('Number of Listings')
plt.xlim(0, 1000) # Limiting x-axis for better readability, as some prices are very high
plt.savefig('price_distribution.png')
plt.show() # Display the plot inline
plt.close()

# 2. Number of Listings per Neighbourhood Group
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='neighbourhood_group', order=df['neighbourhood_group'].value_counts().index, palette='viridis')
plt.title('Number of Listings per Neighbourhood Group')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood Group')
plt.savefig('listings_per_neighbourhood_group.png')
plt.show() # Display the plot inline
plt.close()

# 3. Number of Listings per Room Type
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='room_type', order=df['room_type'].value_counts().index, palette='plasma')
plt.title('Number of Listings per Room Type')
plt.xlabel('Number of Listings')
plt.ylabel('Room Type')
plt.savefig('listings_per_room_type.png')
plt.show() # Display the plot inline
plt.close()

# 4. Price Distribution by Room Type
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='room_type', y='price', palette='coolwarm', showfliers=False) # showfliers=False to hide extreme outliers
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price (USD)')
plt.savefig('price_by_room_type.png')
plt.show() # Display the plot inline
plt.close()

# 5. Geographical Distribution of Airbnb Listings
plt.figure(figsize=(12, 10))
sns.scatterplot(data=df, x='longitude', y='latitude', hue='neighbourhood_group', s=10, alpha=0.6, palette='tab10')
plt.title('Geographical Distribution of Airbnb Listings in NYC')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Neighbourhood Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig('geographical_distribution.png')
plt.show() # Display the plot inline
plt.close()
```
